```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('sonar data.csv',header=None)
```

### see here i am using value_counts to find out how many of the data have mines(M) and rocks(R)
### i am choosing 60th column beacuse the it have the classification of Rocks and Mines
### thing to remember that if i use value_counts --> it will give me all the data of 208 rows
### but if i use value_counts --> it will give me the classified data Rocks no. of data and  MInes no. of data

```python
df[60].value_counts()
```

```python
df.groupby(60).mean()
```

### now i will seperate the labels and the data from the table

```python
x=df.drop(columns=60,axis=1)
```

```python
print(x)
```

```python
y=df[60]
```

```python
print(y)
```

we have give 10% of data to test size, statify Y because it contains the classification between rock and mine

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.1,stratify=y,random_state=1)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(x_train)
print(y_train)
```

## Model training --> Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

### Model Evaluation

```python
#accuracy on testing data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print("Accuracy_score_on_training_data : "  , training_data_accuracy)
```

```python
#accuracy on training data
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print("Accuracy_score_on_testing_data : "  , testing_data_accuracy)
```
